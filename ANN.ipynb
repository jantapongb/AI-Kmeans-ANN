```
import pandas as pd
import pandas as testingpd
from google.colab import drive

drive.mount('/content/drive')

# Assign colum names to the dataset
names = ['SumOfReuestPerSession', 'SumOfSizePerSession', 'RequestRatePerSession', 'FreqOfVisitWebpages', 'Class']

# Read dataset to pandas dataframe
irisdata =pd.read_csv('/content/drive/Shared drives/PROFRESSIONAL_FIRST/DOCTOR/TERM 2/AI/COLAP/AllTrainDataNormalAbnormalANN.csv')
irisdata.head()

# Assign data from first four columns to X variable
X = irisdata.iloc[:, 0:4]

# Assign data from first fifth columns to y variable
y = irisdata.select_dtypes(include=[object])

y.head()
y.Class.unique()

from sklearn import preprocessing
le = preprocessing.LabelEncoder()

y = y.apply(le.fit_transform)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.9)

#---------------------------------------------- Testing Dataset ---------------------------------------------
TestingDS=testingpd.read_csv('/content/drive/Shared drives/PROFRESSIONAL_FIRST/DOCTOR/TERM 2/AI/COLAP/TestingDataSetColapANNMix.csv')
#TestingDS.head()
TestX = TestingDS.iloc[:, 0:4]
TestY = TestingDS.select_dtypes(include=[object])
#TestY.head()
TestY.Class.unique()
Testle = preprocessing.LabelEncoder()
TestY = TestY.apply(Testle.fit_transform)

#----------------------------------------------- Feature Scaling --------------------------------------------
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
TestX = scaler.transform(TestX)

#--------------------------------------------- Training and Predictions ------------------------------------
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)
mlp.fit(X_train, y_train.values.ravel())

predictions = mlp.predict(TestX)
print(predictions)

#---------------------------------------------- Evaluating the Algorithm ---------------------------------
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(TestY,predictions))
print(classification_report(TestY,predictions))
```
